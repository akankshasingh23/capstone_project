### columns having outliers

```python
df.describe(percentiles=[0.01,0.02,0.03,0.97,0.98,0.99]).T
```

```python
def outlier_handle(x):
    x = x.clip(upper=x.quantile(0.98))
    x = x.clip(lower=x.quantile(0.02))
    return x
```

```python
outlier_cols = ['funding_rounds','avg_participants','relationships']
df[outlier_cols] = df[outlier_cols].apply(outlier_handle)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,
                                                 random_state=32)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn import model_selection

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
```

```python
def gen_metrics(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test,y_pred)
    print('Prediction\n',y_pred)
    print('Training score',train_score)
    print('Testing score',test_score)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))
```

### model-1 Decision tree

```python
m1 = DecisionTreeClassifier(criterion='gini',max_depth=7,min_samples_split=20)
gen_metrics(m1,x_train,x_test,y_train,y_test)
```

```python
ypred_m1 = m1.predict(x_test)
```

```python
from sklearn.tree import plot_tree
```

```python
fn = x_train.columns
cn = ['1','0']

plt.figure(figsize=(40,38))
plot_tree(m1,feature_names=fn,class_names=cn,filled=True)
plt.show()

```

### model-2 Random Forest

```python
m2 = RandomForestClassifier(n_estimators=120,criterion='entropy',
                            max_depth=7,min_samples_split=20)
gen_metrics(m2,x_train,x_test,y_train,y_test)

ypred_m2= m2.predict(x_test)
```

```python
plt.figure(figsize=(40,38))
for i in range(1,5):  # i=1,2,3,4
    plot_tree(m2.estimators_[i],feature_names=fn,class_names=cn,filled=True)
    plt.title(f'Decision Tree {i}')
    plt.show()

```

### model - 3 ( LogisticRegression)

```python
m3 = LogisticRegression(max_iter=1000)
m3.fit(x_train,y_train)
ypred_m3= m3.predict(x_test)
gen_metrics(m3,x_train,x_test,y_train,y_test)
```

### model- 4 (Adaboost classifier)

```python
dt1 = DecisionTreeClassifier(max_depth=7)
m4 = AdaBoostClassifier(dt1,n_estimators=100,learning_rate=1)
gen_metrics(m4,x_train,x_test,y_train,y_test)
ypred_m4= m4.predict(x_test)
```

### model-5 (Naive bayes)

```python
m5 = GaussianNB(priors=[0.50,0.50])
gen_metrics(m5,x_train,x_test,y_train,y_test)
ypred_m5= m5.predict(x_test)
```

### from all the above models adaboost classifier is giving the best results
